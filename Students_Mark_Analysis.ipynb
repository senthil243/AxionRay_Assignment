```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis = 1)
df.head()
```

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("5 - 10","6 - 10")
df.head()
```

```python
df.head()
```

**Gender Distribution**

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df,x = "Gender",color="pink",)
ax.bar_label(ax.containers[0])
plt.title("Gender Distrubution")
plt.show()
```

**From the above chart analyzed **

```python
#In the school female is more then the male
```

**Parent education how affect the children mark**

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
#Above result we can show in the heat map
sns.heatmap(gb,annot=True)
plt.title("Parent Edcution vs Student Marks")
plt.show()
```

```python
#from the above analysis parent educ is highly affected on childerns mark
```

**let we check for parent martial status also **

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Parent Marital Status vs Student Marks")
plt.show()
```

```python
#form the above analysis parent marital status is not affecting his or him childernes marks in the school.
```

**Pie chart for EthnicGroup**

```python
uni = df["EthnicGroup"].unique()
print(uni)
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A","group B","group C","group D", "group D"]
plt.title("Distribution Ethnic Group")
plt.pie([groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]],labels= l ,autopct="%1.2f%%" )
plt.show()

```
